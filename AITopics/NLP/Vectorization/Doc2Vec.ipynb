```python
from gensim.models.doc2vec import Doc2Vec, TaggedDocument

# Sample Data
data = ["The cat sat on the mat.", "I love machine learning.", "Python is a great language."]

# Step 1 & 2: Tokenize and Tag
tagged_data = [TaggedDocument(words=doc.lower().split(), tags=[str(i)]) for i, doc in enumerate(data)]
print(tagged_data)
```

```python
model = Doc2Vec(vector_size=20, min_count=1, epochs=50)
model.build_vocab(tagged_data)
model.train(tagged_data, total_examples=model.corpus_count, epochs=model.epochs)

# Step 5: Infer a vector for new text
new_vector = model.infer_vector(["cats", "on", "mats"])
```
